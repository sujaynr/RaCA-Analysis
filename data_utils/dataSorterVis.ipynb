```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
directory = "/Users/sujaynair/Documents/RaCA-Analysis/newMinerals"

spectra_data = {}


for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        parts = filename.split('_')
        element_name = parts[2]  
        
        if element_name not in spectra_data:
            spectra_data[element_name] = []
        
        with open(os.path.join(directory, filename), 'r') as file:
            next(file)
            spectra_values = [float(line.strip()) for line in file]
            spectra_data[element_name].append((filename, spectra_values))

first_element_data = next(iter(spectra_data.values()))[0][1]
x_axis = np.linspace(0.3, 2.5, num=len(first_element_data))

for element_name, spectra_list in spectra_data.items():
    spectra_list.sort(key=lambda x: x[0])
    for filename, spectra in spectra_list:
        plt.figure(figsize=(10, 6))
        plt.plot(x_axis, spectra, label=f'{filename} spectrum')
        plt.title(f'{element_name}')
        plt.xlabel('Wavelength (m)')
        plt.ylabel('Reflectance')
        plt.legend()
        plt.show()
```
